## 2. Ristiintaulukointia
Tiedostossa https://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/kysely.csv on kuvitteellisen henkilöstökyselyn tuloksia.

Dataan on koodattu koulutus-kenttään numeroilla:  
* 1 = peruskoulu
* 2 = toisen asteen koulutus
* 3 = korkeakoulu
* 4 = ylempi korkeakoulu

TyytJohto ja TyytTyöteht -kentät kuvaavat tyytyväisyyttä johtoon ja työtehtäviin asteikolla 1-5.   

työterveys, liikunta ja lomaosake-kentät sisältävät ykkösen jos vastaaja on käyttänyt näitä palveluita, muuten NaN

Ristiintaulukoi/pivotoi/koosta seuraavat jutut:  
* lukumäärät siten, että riveillä perhesuhteet, sarakkeissa tyytyväisyys johtoon (lukumäärät, eli kuinka monta perheellistä on vastannut "tyytyväisyys johtoon"-kohtaan ykkösen jne.
  
  
* prosenttiosuudet siten, että riveillä koulutus tekstinä (esim peruskoulu), sarakkeissa tyytyväisyys työtehtäviin ( prosenttiosuudet 1 desimaalilla riveittäin eli esim montako % vain peruskoulun käyneistä vastaa 1 jne.) 
  
  
* keskiarvot tyytyväisyyksistä johtoon ja työtehtäviin siten, että riveillä on ikäluokka 20-29, 30-39, ... 60-69 ja (eri sarakkeissa) tyytyväisyydet johtoon ja työtehtäviin  
  
  
* kuinka monta prosenttia miehistä ja naisista on käyttäneet työterveys, liikunta ja lomaosake-palveluita  
malli tuloksesta
```
          liikunta lomaosake työterveys
sukupuoli                              
mies         24.6%     11.5%      57.4%
nainen       23.8%      9.5%      57.1%
```
```

````python
import pandas as pd

df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/kysely.csv')




print('Tyytyväisyys johtoon \n\n')

print(pd.crosstab(df['perhetilanne'],df['TyytJohto']))      

print('\n\n  Tyytyväisyys työtehtäviin \n\n')

df['koulutus_mj'] = df['koulutus'].map({1:'peruskoulu',2: 'toisen asteen koulutus', 3: 'korkeakoulu'  , 4:'ylempi korkeakoulu'})

# normalize='index' laskee suhteelliset osuudet riveittäin
df3 = pd.crosstab(df['koulutus_mj'], df['TyytTyöteht'], normalize='index').applymap("{:.1%}".format)

print(df3)



print(' \n\n  Keskiarvot tyytyväisyyksistä johtoon ja työtehtäviin \n\n')

ikaluokat = pd.interval_range(20,70,freq=10,closed='left')
df['ikaluokka'] = pd.cut(df['ikä'], ikaluokat, right=False)

# pivot_table laskee oletuksena keskiarvon
print(df.pivot_table(['TyytJohto', 'TyytTyöteht'], index=['ikaluokka']))


print('\n\n palveluita käyttäneet \n\n')

# jos ei-käyttäneet on nollia ja käyttäneet ykkösiä, keskiarvo kertoo käyttäneiden osuuden

df[['työterveys','liikunta', 'lomaosake']] = df[['työterveys','liikunta', 'lomaosake']].fillna(0)
print(df.pivot_table(['työterveys','liikunta', 'lomaosake'], index=['sukupuoli']).applymap("{:.1%}".format))

# tapa2
print('\n\n tapa2 \n\n')

df4 = df.groupby('sukupuoli')['työterveys','liikunta', 'lomaosake'].sum() # määrät
miehet = df['sukupuoli'].value_counts()['mies'] # kokonaismäärät
naiset = df['sukupuoli'].value_counts()['nainen']  # kokonaismäärät

df4.loc['mies'] = df4.loc['mies']/miehet   
df4.loc['nainen'] = df4.loc['nainen']/naiset
print(df4.applymap("{:.1%}".format))
````

## 3. Liikennelaskentaa

Tiedostossa
http://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/lam.csv  
on automaattisen liikennelaskennan tietoja yhden kuukauden ajalta yhdestä Espoon mittauspisteestä. 

* rivinro
* päiväys
* suuntakoodi (1 tai 2) 
* ajoneuvoluokka
* tuntiväli
* määrä

Koosta/muokkaa taulukosta DataFrame, jossa on 24 riviä, yksi vuorokauden jokaiselle tunnille ja 30 saraketta kuukauden päiville, siten että DataFramessa on kyseisen tunnin ajoneuvomäärien summa (molemmat suunnat, kaikki ajoneuvoluokat). 

Mallia (otsikot voi vaihdella):  
![lam1](https://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/malli1.png)

Laske sitten loppuun sarakkeet:
* suurin, joka kertoo kunkin tunnin suurimman ajoneuvomäärän 30 päivän joukosta
* sija, joka kertoo päivän 1 (1.11.2019) sijoituksen ko. tunnin ajoneuvomäärien suhteen 30 päivän joukossa (eniten = 1.)  
    (siis jos 00-01 -tunneista 1.11. on enemmän kuin minään muuna päivänä, sija = 1. Jos vähemmän kuin muina päivinä, sija = 30)

![lam2](https://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/malli2.png)

````python
import pandas as pd
df = pd.read_csv('http://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/lam.csv')
df['Pvm'] = df['Pvm'].str[:2].astype(int) # päivä otettu omaksi kentäksi

# groupby kahdella ryhmittelykriteerillä antaa Seriesin, jolla monitasoinen indeksi, unstack muuttaa dataframeksi
df2 = df.groupby(['tuntiväli','Pvm'])['määrä'].sum().unstack()


suurin = df2.max(axis=1)

# suurin-saraketta ei ole vielä dataframessa, jotta se ei vaikuta sijaan
# rank antaa jokaiselle solulle sijan rivillä, [1] lopussa antaa sarakkeen 1 arvon (sarakeotsikko 1 = 1. päivä)
sija = df2.rank(ascending=False, axis=1, method='average')[1]

# lisätään dataframeen
df2['suurin'] = suurin
df2['sija'] = sija

df2
````

## 5. Jalkapalloa
Tiedostossa https://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/epl4.txt  on Englannin Valioliigan ottelutulokset kauden 2018-2019 loppuun. Käsittelyn helpottamiseksi siinä on laskettu valmiiksi koti- ja vierasjoukkueiden saamat pisteet ottelusta (3, 1 tai 0).

Tiedostossa https://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/sijat.txt on joukkueiden loppusijoitukset kausittain. 

Ensimmäisestä tiedostosta pitäisi saada koostettua DataFrame, jossa on tieto **kuinka monta pistettä kukin joukkue on saanut kunkin kauden 15 ensimmäisestä ottelustaan**

Toista tiedostoa hyväksi käyttäen pitäisi sitten etsiä vastaus kysymykseen **mitkä ovat olleet suurimmat erot (eniten jäljessä) neljänteen sijaan 15 ottelun jälkeen joukkueilla, jotka ovat sijoittuneet kauden lopussa neljän parhaan joukkoon?**

* laske jokaiselle joukkueelle jokaiselle kaudelle kuinka paljon pisteitä on kauden 15 ensimmäisestä pelistä
* laske jokaiselle kaudelle paljonko on neljänneksi suurin pistemäärä (15 ensimmäisen pelin pistemääristä)
* laske jokaiselle joukkueelle jokaiselle kaudelle paljonko ero on ollut tähän neljänteen pistemäärään
* suodata mukaan vain ne joukkueet, jotka ko. kaudella ovat loppusijoituksissa 4 parhaan joukossa ja laijittele näistä eroista suurimmat (eniten 4 sijaa jäljessä olleet).

Alkua 1. vaiheesta (15 pelin pistemääristä):  

```
Season       1993-1994  1994-1995  1995-1996  1996-1997  1997-1998  1998-1999
Team                                                                            
Arsenal           26.0       19.0       28.0       31.0       27.0       25.0   
Aston Villa       29.0       13.0       27.0       24.0       18.0       30.0   
Barnsley           NaN        NaN        NaN        NaN       13.0        NaN   
Birmingham         NaN        NaN        NaN        NaN        NaN        NaN

Vastaus kysymykseen:  
```
 		 	
Season 	     Team          ero_neloseen  loppusija
2001-2002 	  Man United    6 	        3
2012-2013 	  Arsenal       5 	        4
2017-2018 	  Tottenham     4 	        3
2000-2001 	  Leeds 	    4 	        4
2014-2015 	  Arsenal       4 	        3
```

*Ottelusiirtojen yms. vuoksi joukkeet eivät ole pelanneet samaa tahtia, joten "ero 4. sijaan 15 pelin jälkeen" ei ole ollut välttämättä minkään päivän sarjataulukon tilanne (eri joukkueet ovat saaneet 15 peliä pelattua eri aikoina), vaan tässä lasketaan vain jokaiselta joukkueelta sen 15 ensimmäistä ottelua kultakin kaudelta*.

````python
import pandas as pd

df = pd.read_csv('http://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/epl4.txt')

# tässä ratkaisutavassa tehdään erikseen kotijoukkueiden dataframe ja vierasjoukkuieiden dataframe ja yhdistetään ne
# tällöin jokaisesta ottelusta tulee 2 riviä dataframeen, mutta kunkin joukkueen 15 ensimmäiseen otteluun se ei vaikuta

dfkoti = df[['Season','Date','HomeTeam','HomePoints']]
dfkoti.columns = ['Season','Date','Team','Points']

dfvieras = df[['Season','Date','AwayTeam','AwayPoints']]
dfvieras.columns = ['Season','Date','Team','Points']

# yhdistetään koti- ja vierasdataframet ja reindexillä poistetaan tuplaindeksit
df2 = pd.concat([dfkoti,dfvieras], ignore_index=True)
df2.reindex()

# lajitellaan päiväyksen mukaan 
df2.sort_values('Date', inplace=True)

# kun dataframe on lajiteltu päiväyksen mukaan, antaa [:15] aina 15 ekaa ottelua/joukkue/kausi

def ekat15(x):
    return x[:15].sum()


# groupby kahdella kriteerillä tekee Serieksen 2-tasoisella indeksilla, mutta unstackilla saadaan se taulukkomuotoon
# nyt siis agg-funktiolle ekat15 menee parametrina jokainen team-season -yhdistelmä aikajärjestyksessä (koska lajiteltiin pvm mukaan)
df3  = df2.groupby(['Team','Season'])['Points'].agg(ekat15).unstack()


# välitulos printattuna
print(df3.head())

# applylle menee axis=0 :lla jokaisen sarakkeen (kauden) arvot, sortataan ja otetaan 4. suurin
neloset = df3.apply(lambda x: x.sort_values(ascending=False)[3],axis=0)

# neloset on nyt series jossa on joka kauden 4. suurin pistemäärä 15 ekasta ottelusta
print(neloset)

# neloset -Serieksen indeksinä on kaudet 1993-1994... ja df3:n sarakeindeksinä on samat kaudet, ne voidaankiin laskea vähennyslaskuna
dferot = neloset-df3

# tsekataan välitulos
print(dferot)



dfsijat = pd.read_csv('http://student.labranet.jamk.fi/~varpe/datananal2k2020/kerta4/sijat.txt', index_col='Team')

# suodatetaan vain kauden lopussa 4:n joukossa olleet

# dfsijat-dataframessa on samat indeksit kuin dferot (joukkue ja kausi)
dftop4 = dferot[dfsijat<5]


# unstackilla tehdään Series (arvona 15 ottelun jälkeiset erot indeksinä (kausi, team))
# muutetaan kokonaisluvuiksi (dropna:n jälkeen), sortataan, muutetaan dataframeksi jotta saadaan lisättyä loppusija
dftop4b = dftop4.unstack().dropna().astype(int).sort_values(ascending=False).to_frame()
#lisätään loppusija, unstackilla saadaan dfsijat-dataframeen sama kaksitasoinen indeksi
dftop4b['loppusija'] = dfsijat.unstack().dropna().astype(int)
# nimetään sarake paremmin
dftop4b.rename(columns={0:'ero_neloseen'}, inplace=True)

dftop4b.head(20)
````
